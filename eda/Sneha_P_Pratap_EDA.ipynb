```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

customers = pd.read_csv('C:/Users/sneha/OneDrive/Desktop/Zeotap-Data-Science-Assignment/data/Customers.csv')
products = pd.read_csv('C:/Users/sneha/OneDrive/Desktop/Zeotap-Data-Science-Assignment/data/Products.csv')
transactions = pd.read_csv('C:/Users/sneha/OneDrive/Desktop/Zeotap-Data-Science-Assignment/data/Transactions.csv')

print(customers.info())
print(products.info())
print(transactions.info())

print(customers.isnull().sum())
print(products.isnull().sum())
print(transactions.isnull().sum())

region_counts = customers['Region'].value_counts()
region_counts.plot(kind='bar', title='Transactions by Region')
plt.show()

top_products = transactions.groupby('ProductID')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', title='Top 10 Products')
plt.show()

merged = transactions.merge(customers, on='CustomerID').merge(products, on='ProductID')

sales_by_category = merged.groupby('Category')['TotalValue'].sum()
sales_by_category.plot(kind='pie', autopct='%1.1f%%', title='Sales by Category')
plt.show()


```
